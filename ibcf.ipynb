```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dish_df = pd.read_csv('./data/HCAT Data Set - Sheet2.csv')
rating_df = pd.read_csv('./data/HCAT Data Set - Rating 2.csv')
```

```python
dish_df
```

```python
dishes = list(dish_df['dish_id'])
```

```python
rating_df
```

```python
ratings_matrix = rating_df.pivot(index='User ID', columns='Dish ID', values='Rating')
ratings_matrix
```

```python
rated_dishes = list(rating_df['Dish ID'].unique())
unrated_dishes = [dish for dish in dishes if dish not in rated_dishes]
unrated_dishes
```

```python
new_columns = pd.DataFrame(np.nan, index=ratings_matrix.index, columns=unrated_dishes)
ratings_matrix = pd.concat([ratings_matrix, new_columns], axis=1)
```

```python
ratings_matrix
```

```python
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix
```

```python
dishes_matrix = ratings_matrix.T
dishes_matrix
```

```python
cosine_sim = cosine_similarity(dishes_matrix)
cosine_sim_df = pd.DataFrame(cosine_sim, index=dishes_matrix.index, columns=dishes_matrix.index)
cosine_sim_df
```

```python

N = 3  # Number of nearest neighbors (you can change this)

# Step 2: Create an empty matrix to store predicted ratings
predicted_ratings_matrix = ratings_matrix.copy()

# Step 3: Iterate over each user and each dish
for user in ratings_matrix.index:
    for dish in ratings_matrix.columns:
        # Only calculate for unrated dishes (where the rating is 0 or NaN)
        if ratings_matrix.at[user, dish] == 0:
            
            # Step 4: Get the similarities of the current dish with all other dishes
            similarities = cosine_sim_df[dish].drop(dish)  # Drop the self-similarity (diagonal)
            
            # Step 5: Find the top N most similar dishes that the user has rated
            rated_dishes = ratings_matrix.loc[user][ratings_matrix.loc[user] > 0]  # Dishes rated by the user
            similarities_rated_dishes = similarities[rated_dishes.index]  # Similarities to rated dishes
            
            # Step 6: Get the top N similar dishes
            top_n_similar_dishes = similarities_rated_dishes.nlargest(N)
            
            # Step 7: Calculate the weighted average for the prediction
            numerator = np.sum(top_n_similar_dishes * rated_dishes[top_n_similar_dishes.index])
            denominator = np.sum(np.abs(top_n_similar_dishes))
            
            # Step 8: Store the predicted rating
            if denominator != 0:  # Avoid division by zero
                predicted_ratings_matrix.at[user, dish] = numerator / denominator
            else:
                predicted_ratings_matrix.at[user, dish] = 0  # If no similar dishes, set prediction to 0 (or use another strategy)

# Step 9: Remove already-rated dishes (set them to NaN or 0)
for user in predicted_ratings_matrix.index:
    rated_dishes_by_user = ratings_matrix.loc[user][ratings_matrix.loc[user] > 0].index
    predicted_ratings_matrix.loc[user, rated_dishes_by_user] = np.nan  # or 0 if you prefer

# Display the predicted ratings matrix
predicted_ratings_matrix
```

```python
recommendations = {}

for user in predicted_ratings_matrix.index:
    # Step 2: Filter dishes with predicted ratings > 0
    filtered_ratings = predicted_ratings_matrix.loc[user][predicted_ratings_matrix.loc[user] > 0]
    
    # Step 3: Sort ratings in descending order
    top_rated_items = filtered_ratings.sort_values(ascending=False)
    
    # Step 4: Select the top 3 dishes
    top_3_items = top_rated_items.head(3).index.tolist()
    
    # Step 5: Store the recommendations for each user
    recommendations[user] = top_3_items

# Display the recommendations
for user, items in recommendations.items():
    print(f"User {user}: Top 3 Recommended Dishes -> {items}")
```
